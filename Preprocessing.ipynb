```python
import pandas as pd
import numpy as np

```

```python
crime_df = pd.read_csv('DATA/crimes.csv' , encoding = 'windows-1252') 

crime_df.dtypes

# crime_df.head()
```

```python
# crime_df.info()
```

```python
# fill empty SHOOTING column with 'N'
crime_df['SHOOTING'].fillna('N', inplace=True)

# convert OCCURRED_ON_DATE to datetime
crime_df['OCCURRED_ON_DATE'] = pd.to_datetime(crime_df['OCCURRED_ON_DATE'])


crime_df.head()
```

```python
# drop Offence_DEscription column
crime_df.drop('OFFENSE_DESCRIPTION', axis=1, inplace=True)

crime_df.head()
```

```python
crime_df.dtypes
```

```python
# # save to postgresql
# from sqlalchemy import create_engine
# engine = create_engine('[db_type]://[username]:[password]@[host]:[port]/[database_name]')
# engine = create_engine('postgresql://amrit:1234@localhost:5432/crimes_in_boston')
# crime_df.to_sql('crime', engine, if_exists='replace', index=False)

```

```python
# for  non-repeating incident number  add _0 as suffix
# for Incident number I162030584  there are 13 rows so I162030584_0 ,I162030584_1 ...... and so on 

crime_df['INCIDENT_NUMBER'] = crime_df['INCIDENT_NUMBER'].astype(str) + '_' + crime_df\
    .groupby('INCIDENT_NUMBER')\
    .cumcount().astype(str)

    #cumcount() returns the number of rows before the current row, starting from 0

# here we are adding suffix to INCIDENT_NUMBER column  so that it becomes unique

```

```python

crime_df['INCIDENT_NUMBER'].value_counts()  # verfiy if there are any duplicate incident number
```

```python
# check for incident number I162030584 with differnt suffix
crime_df[crime_df['INCIDENT_NUMBER'] == 'I162030584_12']
```
